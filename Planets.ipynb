```python
path_planet = '/content/drive/MyDrive/planets.csv'
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv(path_planet)
```

```python
df.head()
```

```python
# @title Number of Discoveries by Method and Year
# Create a figure and axes
plt.figure(figsize=(10, 6))
ax = plt.gca()
sns.barplot(x='year', y='number', hue='method', data=df)
plt.xlabel('Year')
plt.xticks(rotation=45)
_ = plt.ylabel('Number of Discoveries')
```

```python
# prompt: Using dataframe df: box plot

import altair as alt
alt.Chart(df).mark_boxplot().encode(
    x = 'method',
    y = 'mass'
).properties(height=400, width=600)
```

```python
# @title mass

from matplotlib import pyplot as plt
df['mass'].plot(kind='line', figsize=(8, 4))
plt.xlabel('Mass (Jupiter Masses)')
plt.ylabel('Frequency')
plt.title('Distribution of Exoplanet Masses')
plt.gca().spines[['top', 'right']].set_visible(False)
```
